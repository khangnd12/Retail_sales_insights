# Online Retail Data Audit

This notebook will inspect the raw retail dataset and identify data-quality
issues before cleaning and analysis.

```python
import pandas as pd
from pathlib import Path

print(f"Pandas version: {pd.__version__}")
print("Notebook environment is working.")
```

```python
data_path = Path("../data/raw/Online Retail.xlsx")

df = pd.read_excel(data_path)

print("Dataset loaded successfully.")
```

```python
df.head()
```

```python
df.sample(5)
```

```python
rows, columns = df.shape

print(f"Rows: {rows:,}")
print(f"Columns: {columns}")
```

The dataset contains many transaction lines and several fields describing
the invoice, product, customer, price, quantity, date, and country.

```python
df.columns.tolist()
```

| Column      | Meaning                       |
| ----------- | ----------------------------- |
| InvoiceNo   | Invoice or transaction number |
| StockCode   | Product code                  |
| Description | Product name                  |
| Quantity    | Number of units               |
| InvoiceDate | Date and time of transaction  |
| UnitPrice   | Price per unit                |
| CustomerID  | Customer identifier           |
| Country     | Customer country              |


```python
df.dtypes
```

```python
df.describe(include="all")
```

```python
missing_report = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_percentage": df.isna().mean() * 100,
})

missing_report = missing_report.sort_values(
    "missing_percentage",
    ascending=False,
)

missing_report
```

The customerID and Description cols contain missung values, with around 25% and 0,27% respectively

```python
duplicate_count = df.duplicated().sum()

print(f"Duplicate rows: {duplicate_count:,}")
```

```python
df[df.duplicated(keep=False)].head(10)
```

Identified exact duplicate records, rather than genuine separate transactions

```python
df["Quantity"].describe()
```

```python
print("Positive quantities:", (df["Quantity"] > 0).sum())
print("Zero quantities:", (df["Quantity"] == 0).sum())
print("Negative quantities:", (df["Quantity"] < 0).sum())
```

```python
df[df["Quantity"] < 0].head(10)
```

These negative quantities may represent the returned product

```python
df["UnitPrice"].describe()
```

```python
print("Zero or negative prices:", (df["UnitPrice"] <= 0).sum())
```

```python
df[df["UnitPrice"] <= 0].head(10)
```

Detected negative UnitPrice values, these rows may need special handling during cleaning

```python
cancelled_mask = (
    df["InvoiceNo"]
    .astype(str)
    .str.startswith("C")
)

cancelled_orders = df[cancelled_mask]

print(f"Cancelled transaction rows: {len(cancelled_orders):,}")
```

```python
cancelled_orders.head(10)
```

```python
df["Country"].value_counts().head(10)
```

```python
country_share = (
    df["Country"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

country_share.head(10)
```

The most dominant country is The UK

```python
print("Unique invoices:", df["InvoiceNo"].nunique())
print("Unique products:", df["StockCode"].nunique())
print("Unique customers:", df["CustomerID"].nunique())
print("Unique countries:", df["Country"].nunique())
```

##Initial Data-quality checking:
1. The dataset contains 541,909 rows and 8 columns
2. Missing Value may appear in CostumerID and Description columns
3. Some transaction quantities are negative, which may represent the returned product
4. Some unit prices are zero or negative
5. Some InvoiceNo begin with C, which may indicate cancelled transactions
6. Duplicate rows require further investigation
7. Rows without customer IDs may still be useful for sales analysis, but they cannot be used for customer segmentation.
